# Centanet URL extraction

```python
import os
from dotenv import load_dotenv, find_dotenv

load_dotenv(find_dotenv()) 
```

```python
# centanent url the last page
# page1
url_sale = os.getenv("url_sale")
url_rent = os.getenv("url_rent")
executable_path = os.getenv("executable_path")


service = Service(executable_path)
options = webdriver.ChromeOptions()
options.add_argument('--disable-blink-features=AutomationControlled')

driver = webdriver.Chrome(service=service, options=options)

url_list = [url_sale,url_rent] #[sales, rent]
for url in url_list:
    if url_list[0]==url:
        flg = 'sale'
    else:
        flg = 'rent'
    driver.get(url)
    page = 0
    for page in range(417):
        page_source = driver.page_source
        element = driver.find_element(By.CSS_SELECTOR, "li.number.active")
        active_page_value = element.get_attribute("textContent")
        with open("../pages/"+flg+"{:03d}".format(int(active_page_value))+".html", "w", encoding="utf-8") as f:
            f.write(page_source)
        print(flg," page: ",active_page_value)
        button = driver.find_element("xpath","//i[@class='el-icon el-icon-arrow-right']") 
        time.sleep(random.uniform(1,3)) 
        driver.execute_script("arguments[0].click();", button) 
        page=+1
driver.quit()



```
